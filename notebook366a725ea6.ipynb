```python
"""EV Adoption Trends: Analyze the adoption rate of electric vehicles over the years.
How has the demand for EVs changed over time"""
    
import pandas as pd

file_path = '/kaggle/input/india-ev-market-data/ev_sales_by_makers_and_cat_15-24.csv' 
ev_data = pd.read_csv(file_path)

years = [str(year) for year in range(2015, 2025)] 
total_sales_over_year = ev_data[years].sum()

# Converting result to DataFrame
t_sale = total_sales_over_year.reset_index()
t_sale.columns = ['Year', 'Total Sales']

print(t_sale)
output_file_path = '/kaggle/working/EV_Adoption_Trends.xlsx'
t_sale.to_excel(output_file_path, index=False)

print(f"EV Adoption Trends saved to {output_file_path}")
```

```python
"""Manufacturer Performance: Compare the performance of different EV manufacturers. Which 
companies have the highest market share,and what factors contribute to their success"""

import pandas as pd

file_path = '/kaggle/input/india-ev-market-data/ev_sales_by_makers_and_cat_15-24.csv'
ev_data = pd.read_csv(file_path)

years = [str(year) for year in range(2015, 2025)] 

# Calculating total sales per manufacturer over the entire period
ev_data['Total Sales'] = ev_data[years].sum(axis=1)

m_sales = ev_data.groupby('Maker')['Total Sales'].sum().reset_index()

# market share for each manufacturer
total_market_sales = m_sales['Total Sales'].sum()
m_sales['Market Share (%)'] = (m_sales['Total Sales'] / total_market_sales) * 100

sort = m_sales.sort_values(by='Market Share (%)', ascending=False)

print(sort)
output_file_path = '/kaggle/working/Manufacturer_Performance.xlsx'
sort.to_excel(output_file_path, index=False)

print(f"Manufacturer Performance saved to {output_file_path}")
```

```python
"""Price vs. Features: Investigate the relationship between the price of EVs and their 
features. Is there a correlation between price and key attributes such as battery life, 
range, or charging time"""

import pandas as pd
data = {
    'Price': [350000, 480000, 620000, 850000, 700000, 1100000],  
    'Battery_Life': [3000, 5000, 7000, 600, 400, 800],            
    'Range': [280, 450, 550, 680, 540, 760],                    
    'Charging_Time': [10, 5.5, 4.5, 3, 4, 2]   
}
df = pd.DataFrame(data)
correlation_matrix = df.corr()

print(correlation_matrix)
output_file = '/kaggle/working/EV_Price_vs_Features_Correlation.xlsx'
correlation_matrix.to_excel(output_file)

print(f'Price vs. Features correlation data saved to {output_file}')
```
